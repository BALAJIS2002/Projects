```python
import numpy as np
import pandas as pd
```

```python
# Set a random seed for reproducibility
np.random.seed(42)
# Generate synthetic data
num_samples = 1000
# Features: DL,ML,CN,MV,DBS
DL_scores = np.random.randint(60, 101, num_samples)
ML_scores = np.random.randint(60, 101, num_samples)
CN_scores = np.random.randint(60, 101, num_samples)
MV_scores = np.random.randint(60, 101, num_samples)
DBS_scores = np.random.randint(60, 101, num_samples)

# Calculate the average score
average_scores = (DL_scores + ML_scores + CN_scores+ MV_scores+DBS_scores) // 5

# Create a DataFrame
data = pd.DataFrame({
    'DL': DL_scores,
    'ML': ML_scores,
    'CN': CN_scores,
    'MV': MV_scores,
    'DBS': DBS_scores,
    'Average': average_scores
})
grades = ['A', 'B', 'C', 'D']
# Define a grading function
def assign_grades(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
# Assign grades based on average scores
data['Grade'] = data['Average'].apply(assign_grades)
# Display the first few rows of the dataset
print(data.head())
# Save the dataset to a CSV file
data.to_csv('student_marks_dataset.csv', index=False)
```

```python
data.describe()
```

```python
# Perform one-hot encoding on the 'Grade' column
y = pd.get_dummies(data['Grade'], prefix='Grade')
y.head()
```

```python
# Drop the 'Grade' column to create the DataFrame x
x=data.drop(["Grade"],axis=1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, InputLayer

# Assuming 'grade' is a column in your dataset
unique_grades = len(set(data['Grade']))

model = Sequential()
model.add(InputLayer( input_shape=len(x.columns,)))
model.add(Dense(units=256, activation='relu'))  # Assuming input features are in x
model.add(Dense(units=512, activation='relu'))
model.add(Dense(units=1024, activation='relu'))
model.add(Dense(units=2048, activation='relu'))
model.add(Dense(units=512, activation='relu'))
model.add(Dense(units=1024, activation='relu'))
model.add(Dense(units=2048, activation='relu'))
model.add(Dense(units=unique_grades, activation="softmax"))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train,y_train,epochs=150,batch_size=64, shuffle=True)
```

```python
score = model.evaluate(x_test,y_test)
print("accuracy",score)
```

```python
i = model.predict([[91,92,95,96,55,85]])
idx = np.argmax(i)
print(f"The estimated grade for the following marks is {grades[idx]} with accuracy of {i[0][idx]*100}%")
```
